```python
# Que1. Load the given retail sales dataset into a Pandas DataFrame and display
# the first 5 rows along with the dataset information to understand its
# structure

import pandas as pd
import numpy as np

df = pd.read_csv("Python retail_dataset.csv")

print(df.head())

print(df.info())
```

```python
# Q2. Inspect the dataset and identify missing values in each column. Display
# the total number of missing values present in the dataset.

print(df.isnull().sum())

print("Total missing values:", df.isnull().sum().sum())
```

```python
# Q3. Replace missing values in the Quantity and Price columns with their
# respective mean values using Pandas
```

```python
# solution 3
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())

print("Missing values after replacement:")
print(df[['Quantity','Price']].isnull().sum())
```

```python
# Q4. Remove all rows where the Product Category or Region column contains
# missing values.

print("Rows before cleaning:", df.shape[0])

df = df.dropna(subset=['Product Category','Region'])

print("Rows after cleaning:", df.shape[0])
```

```python
# Q5. Create a new column called Revenue using NumPy, where:
# Revenue = Quantity × Price
# Display the updated dataset with the new column

df['Revenue'] = df['Quantity'] * df['Price']

print("Dataset after adding Revenue column:\n")
print(df[['Quantity','Price','Revenue']].head())
```

```python
# Q6. Using NumPy, calculate the total revenue generated from all transactions
# in the dataset.

# Q6 Calculate total revenue using NumPy

import numpy as np

total_revenue = np.sum(df['Revenue'])

print("Total Revenue generated from all transactions is:")
print(total_revenue)
```

```python
# Q7. Group the dataset by Product Category and calculate the total revenue
# generated by each category.


category_revenue = df.groupby('Product Category')['Revenue'].sum()

print("Total Revenue generated by each Product Category:\n")
print(category_revenue)
```

```python
# Q8. Based on the revenue calculated for each product category, identify the
# top 3 categories with the highest revenue and the bottom 3 categories
# with the lowest revenue.



top3 = category_revenue.sort_values(ascending=False).head(3)
bottom3 = category_revenue.sort_values().head(3)

print("Top 3 Categories with Highest Revenue:\n")
print(top3)

print("\nBottom 3 Categories with Lowest Revenue:\n")
print(bottom3)
```

```python
# Q9. Group the dataset by Region and calculate the total revenue generated in
# each region. Identify the region with the highest revenue and the region with
# the lowest revenue.


region_revenue = df.groupby('Region')['Revenue'].sum()

highest_region = region_revenue.idxmax()
lowest_region = region_revenue.idxmin()

print("Total Revenue generated in each Region:\n")
print(region_revenue)

print("\nRegion with Highest Revenue:", highest_region)
print("Region with Lowest Revenue:", lowest_region)
```

```python
 # Q10. Extract the month from the Date column using Pandas and group the
  # dataset by month to calculate the total monthly revenue. Also, calculate the
# mean, median, and standard deviation of the Revenue column using NumPy.



df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

monthly_revenue = df.groupby('Month')['Revenue'].sum()

import numpy as np

mean_revenue = np.mean(df['Revenue'])
median_revenue = np.median(df['Revenue'])
std_revenue = np.std(df['Revenue'])

print("Monthly Revenue:\n")
print(monthly_revenue)

print("\nRevenue Statistics:\n")
print("Mean Revenue:", mean_revenue)
print("Median Revenue:", median_revenue)
print("Standard Deviation:", std_revenue)
```
