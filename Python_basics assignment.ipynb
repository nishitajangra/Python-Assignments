# Question 2 : Explain the difference between List and Tuple in Python.
Feature	List	Tuple
Mutability	Mutable (can change elements)	Immutable (cannot change elements)
Syntax	Uses square brackets []	Uses parentheses ()
Performance	Slower	Faster
Example	[1,2,3]	(1,2,3)

# Question 2 : Explain the difference between List and Tuple in Python.
| Feature     | List                          | Tuple                              |
| ----------- | ----------------------------- | ---------------------------------- |
| Mutability  | Mutable (can change elements) | Immutable (cannot change elements) |
| Syntax      | Uses square brackets `[]`     | Uses parentheses `()`              |
| Performance | Slower                        | Faster                             |
| Example     | `[1,2,3]`                     | `(1,2,3)`                          |


# Question 3 : What is a function in Python? Why are functions useful?
A function is a block of reusable code that performs a specific task.

Functions help to:

Reuse code
Reduce repetition
Make programs easier to read and maintain

```python
def add_numbers(a, b):
    result = a + b
    return result

sum = add_numbers(5, 3)
print("The sum is:", sum)
```

```python
# Question 5 : Write a Python program to check whether a number is even or odd using conditional statements.

num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")
```

```python
# Question 6 : Write a program to print numbers from 1 to 10 using a loop.

i = 1

while i <= 10:
    print(i)
    i = i + 1
```

```python
# Question 7 : Create a list of five numbers and print the maximum number from the list.

numbers = [10, 25, 5, 30, 15]

max_number = max(numbers)

print("Maximum number is:", max_number)
```

```python
# Question 8 : Write a Python program to remove duplicate values from a list using a set.

numbers = [1, 2, 2, 3, 3, 4, 4, 5, 6, 5]

unique_numbers = list(set(numbers))

print("List after removing duplicates:", unique_numbers)
```

```python
# Question 9 : Write a function that returns the square of a number.

def square(num):
    return num * num

result = square(25)
print("Square of the number is:", result)
```

```python
# Question 10 : Write a Python program to count how many times a number appears in a list. Example List [2,3,4,2,5,2]

numbers = [2, 3, 4, 2, 5, 2]

num = int(input("Enter the number to count: "))

count = numbers.count(num)

print("The number appears", count, "times in the list.")
```
