**Q3 Customer Purchase Analysis**

    Identify the top 5 customers based on total sales.
Hint: Use GroupBy with aggregation

```python
customer_sales = df.groupby('Customer')['Sales'].sum()        # Group by Customer and calculate total sales

top_customers = customer_sales.sort_values(ascending=False)      # Sort in descending order

top_5_customers = top_customers.head(5)          # Get top 5 customers

print("Top 5 Customers based on Total Sales:")
print(top_5_customers)
```

*insights:*
    Top 5 customers contribute a significant portion of total revenue, indicating customer concentration risk or opportunity.

**Q6 Multi-Level Grouping**

Group the dataset by: Region and Category. 
Calculate the total sales for each combination.
What insights can you derive from this analysis?

```python
# Multi-level grouping
multi_group = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()

print("Total Sales by Region and Category:\n")
print(multi_group)
```

```python
multi_group = multi_group.sort_values(by='Sales', ascending=False)

print("\nSorted Sales (High to Low):\n")
print(multi_group)
```

**INSIGHTS**
  
    -Electronics category performs highest in the East region.
    
    -Office sales are lower in some regions, indicating weaker demand.
    
    -Electronics Supplies show consistent performance across regions.

**Q7 Pivot Table Creation**
Create a pivot table showing:

Rows → Region

Columns → Product

Values → Total Sales

Which region sells the most laptops?

```python
# Create pivot table
pivot_table = df.pivot_table(
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc='sum'
)

print("Pivot Table:\n")
print(pivot_table)
```

```python
#region with highest Laptop sales

print("\nLaptop Sales by Region:\n")
print(pivot_table['Laptop'].sort_values(ascending=False))
```

**INSIGHTS**

    The EAST region has the highest laptop sales, indicating strong demand.
    
    Other regions show comparatively lower sales, suggesting growth opportunities.

**Q8 Profitability Analysis**
    Calculate the average profit for each product category.
Which category is most profitable?

```python
# Average profit per category
category_profit = df.groupby('Category')['Profit'].mean().reset_index()

# Sort descending
category_profit = category_profit.sort_values(by='Profit', ascending=False)

print("Average Profit by Category:\n")
print(category_profit)
```

```python
#Most Profitable Category
top_category = df.groupby('Category')['Profit'].mean().nlargest(1)

print(top_category)
```

MOST PROFITABLE CATEGORY IS : **Office**

**Q9 Discount Impact Study**

Analyze the relationship between discount and profit.
Tasks:

Calculate the average profit for each discount level

Identify whether higher discounts lead to lower profit

```python
# Average profit for each discount level
discount_profit = df.groupby('Discount')['Profit'].mean().reset_index()

# Sort by discount
discount_profit = discount_profit.sort_values(by='Discount')

print("Average Profit by Discount Level:\n")
print(discount_profit)
```

**INSIGHTS**

    The average profit decreases consistently as the discount increases.
    
    At 0% discount, profit is highest (~751), indicating maximum margin.

    Even a small discount (5%) reduces profit significantly.

    At 20% discount, profit drops drastically (~90), showing very low profitability.

    This clearly indicates a strong negative relationship between discount and profit.

**Q10 End-to-End Data Wrangling Task**

Perform the following workflow:

Filter the dataset where Sales > 3000

Group data by Region and Product

Calculate total sales

Reshape the result into a pivot table

Explain the business insights obtained from this analysis

```python
high_sales = df[df['Sales'] > 3000]

print("Filtered Data (Sales > 3000):\n")
print(high_sales.head())
```

```python
grouped_data = high_sales.groupby(['Region', 'Product'])['Sales'].sum().reset_index()

print("\nGrouped Data:\n")
print(grouped_data)
```

```python
pivot_result = grouped_data.pivot(
    index='Region',
    columns='Product',
    values='Sales'
)

print("\nFinal Pivot Table:\n")
print(pivot_result)
```

**Explanation**

    The dataset was filtered to include only high-value transactions (Sales > 3000).

    The filtered data was grouped by region and product to analyze performance.
    
    A pivot table was created to compare product-wise sales across regions.

    This approach helps identify top-performing products in each region.
